# Graph Visualization




Univariant Analysis

* Univariant Analysis of numerical features : Purpose is to analyse distribution of values of single variables.

```python
plt.figure(figsize=(10, 10))
plt.suptitle('Numerical Features univariant analysis', fontsize=15, fontweight='bold', alpha=0.9, y=1)

for i in range(0, len(numerical_column)):
    plt.subplot(3, 3, i+1)
    sns.kdeplot(x=df[numerical_column[i]],shade=True, color='b')
    plt.xlabel(numerical_column[i])
```

```python
plt.figure(figsize=(10, 10))
plt.suptitle('Categorical Features univariant analysis', fontsize=15, fontweight='bold', alpha=0.9, y=1)

for i in range(0, len(categorical_column)):
    plt.subplot(3, 2, i+1)
    sns.countplot(x=df_final[categorical_column[i]]) #,shade=True, color='b')
    plt.xlabel(categorical_column[i])
```

# Converting categorical data into numerical data using mapping function 

```python
df_final.columns
```

* Turning Gender column into integer data

```python
df_final['Gender']= df_final['Gender'].map({'M':1,'F':0})
df_final.Gender.dtype
```

```python
df_final.Age.unique()
```

* Turning Age column into integer data

```python
df_final['Age'] = df_final['Age'].map({'0-17':1, '55+':2, '26-35':3, '46-50':4, '51-55':5, '36-45':6, '18-25':7})
df_final.Age.dtype
```
